```python
altura = int(input('digite a altura da arvore: '))

for i in range(1, altura):
    for _ in range(altura - i):
        print(' ',end='')

    for _ in range(i):
        print('*', end='')

    print()
    
```

```python

lista_usuario = []
while True:

    print(30*'_','menu',30*'_')
    print('1. cadastrar usuario')
    print('2. listar usuario ')
    print('3. excluir usuario')
    print('4. inserir nome em uma posicao ')
    print('5. sair')
    print(60*'_')

    opcao = input('digite a opcao desejada: ')
    if opcao == '1':
        nome = input('digite seu nome aqui: ').lower()
        if nome!= '':
            lista_usuario.append(nome)
            print(f'usuario {nome} cadastrado com sucesso')
        else:
            print('digite algum valor!: ')
 #listar usuario    
    elif opcao == '2':
        for i, n in enumerate(lista_usuario):
            print(f'{i + 1}º {n}')
 #excluir usuario
    elif opcao == '3':

        nome = input('digite o nome que deseja excluir: ').lower()
        for i in lista_usuario:
            if nome == i:
                lista_usuario.remove(i)
                print('usuario removido com sucesso')

#inserir em uma posicao
    elif opcao == '4':
        novo_nome = input('digite o nome que deseja inserir').lower()
        pos_nome = int(input('digite a posicao que deseja inserir: '))
        #correcao da posicao 
        pos_nome -= 1
        if pos_nome >= 0 and pos_nome <= len(lista_usuario):
            lista_usuario.insert(pos_nome, novo_nome)
        else:
            print('posicao invalida')

# sai do sistema
    elif opcao == '5':
        print(lista_usuario)
        break
```
